```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(iris["data"],iris["target"],test_size=0.3,random_state=42)
```

```python
X_train
```

```python
model=DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,8))
tree.plot_tree(model,filled=True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Decision Tree pre-pruning and hyperparameter tuning

```python
param={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,4,5,6,7,9,10],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=param,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_cv_pred=grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_cv_pred))
print(classification_report(y_test,y_cv_pred))
```
