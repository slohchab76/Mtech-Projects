```python
from sklearn.datasets import load_diabetes
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn.model_selection import GridSearchCV
import warnings
```

```python
diabetes=load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
diabetes
```

```python
df_diabetes=pd.DataFrame(diabetes.data,columns=['age','sex','bmi','bp','s1','s2','s3','s4','s5','s6'])
```

```python
df_diabetes
```

```python
df_target=pd.DataFrame(diabetes.target)
```

```python
df_target
```

```python
df_diabetes.info()
```

```python
df_target.info()
y=diabetes['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(df_diabetes,y,test_size=0.3,random_state=42)
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
y_test.head()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(X_train.corr(),annot=True)
```

```python
regressor=DecisionTreeRegressor()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
print(r2_score(y_test,y_pred),mean_squared_error(y_test,y_pred))
```

```python
param={
    
    'criterion': ['squared_error', 'friedman_mse'],
    'max_depth': [2, 3, 4, 5,10,15,20],
    'min_samples_leaf': [5, 10, 20],
    'min_samples_split': [10, 20, 40],
    'splitter':['best','random'],
    'max_features':['auto','sqrt','log2']
}
```

```python
param
```

```python
grid=GridSearchCV(estimator=regressor,param_grid=param,cv=5,scoring='r2')
```

```python
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_cv_pred=grid.predict(X_test)
```

```python
print(r2_score(y_test,y_cv_pred),mean_squared_error(y_test,y_cv_pred))
```
